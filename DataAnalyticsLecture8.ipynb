Correlation Coefficient (r)
The correlation coefficient (denoted as r) ranges from -1 to 1.
A value  is 1 indicates a  perfect positive  correlation.
A value  is -1 indicates a perfect negative  correlation.
A value is 0 indicates no correlation.
Interpretation of Correlation Values
r = 1: Perfect positive correlation.

The variables move in exactly the same direction.
As one variable increases, the other also increases proportionally.
0.7 ≤ r < 1: Strong positive correlation.

There is a strong, direct relationship between the variables.
0.4 ≤ r < 0.7: Moderate positive correlation.

There is a moderate relationship where, as one variable increases, the other tends to increase, but not perfectly.
0 < r < 0.4: Weak positive correlation.

There is a slight positive relationship between the variables.
r = 0: No correlation.

There is no linear relationship between the variables.
-0.4 < r < 0: Weak negative correlation.

There is a slight inverse relationship between the variables.
-0.7 < r ≤ -0.4: Moderate negative correlation.

There is a moderate inverse relationship where, as one variable increases, the other tends to decrease.
-1 < r ≤ -0.7: Strong negative correlation.

There is a strong, inverse relationship between the variables.
r = -1: Perfect negative correlation.

The variables move in exactly opposite directions.
As one variable increases, the other decreases proportionally.
